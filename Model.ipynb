```python
X, y = functions.generate_data(clauses=10)
```

```python
checkpointer = ModelCheckpoint('checkpoint.h5', monitor='accuracy', mode='max', save_best_only=True)
```

```python
model = Sequential()
model.add(LSTM(256, input_shape=np.shape(X)[1:]))
model.add(RepeatVector(np.shape(y)[1]))
model.add(LSTM(128, return_sequences=True))
model.add(TimeDistributed(Dense(len(functions.alphabets), activation='softmax')))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
for i in range(20):
    print(f"\n\n{f'    Epoch: {i+1:>3}    ':-^100}")
    X, y = functions.generate_data(clauses=50_000, randomize=True)
    model.fit(X, y, epochs=10, batch_size=100, callbacks=[checkpointer])
```

```python
X_test, y_test = functions.generate_data(clauses=20)
```

```python
X_test_str = functions.tensor2text(X_test)
y_test_str = functions.tensor2text(y_test)
pprint(X_test_str)
```

```python
y_pred = functions.predict(X_test_str, model)
```
